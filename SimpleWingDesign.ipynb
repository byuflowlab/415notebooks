# A more general geometry (non-interactive)
![wing parameters](wing2.png)

The wing can be made up of an arbitrary number of linear segments.  

Inputs:

- $xle, yle, zle$: define the points along the leading edge (note that the use of z allows you model winglets, polyhedral, etc.).  
- $c$: chord
- $\theta$: twist in radians.  
- $npanels$: number of panels between segments.  The length of the npanels array will be one less than the others.  Note that if npanels is of length 1 it still needs to be written as an array.  I'd recommend about 100 total panels across the wing.

Outpus:

- The induced drag coefficient is inviscid only.  
- The moment coefficient is about the quarter chord of the root section.  
- The reference area is the gross planform area.  
- The reference chord is the mean aerodynamic chord.

```julia

# ----- inputs ------
xle = [0.0 1.4 4.0 4.2]
yle = [0.0 4.5 10 10]
zle = [0.0 0.0 0.0 2]
chord = [4.8 2.8 1.2 1.0]
theta = [2.0 1.0 -1.0 0.0]*pi/180
npanels = [40 50 10]
alpha = 9.0*pi/180  # angle of attack
clmax = 1.3  # clmax just as a reference line
# --------------------

wing(xle, yle, zle, chord, theta, npanels, alpha, clmax)


```

# Biplane

one new parameter: 

- voffset: vertical offset between two identical wings

Uses same reference area as if it was a single wing (i.e., the area is not doubled).

```julia
# ----- inputs ------
xle = [0.0 1.4 4.0]
yle = [0.0 4.5 10]
zle = [0.0 0.0 0.0]
chord = [4.8 2.8 1.2]
theta = [2.0 1.0 -1.0]*pi/180
npanels = [40 50]
alpha = 9.0*pi/180  # angle of attack
clmax = 1.3  # clmax just as a reference line
# --------------------

voffset = 2

biplane(xle, yle, zle, chord, theta, npanels, voffset, alpha, clmax)

```
